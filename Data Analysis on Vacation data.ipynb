## EDA = Exploratory Data Analysis ##

It is a method utilized to convert data into the information.

1. `Univariate Analysis`: When we take single column for analysis, it is called univariate analysis.
    - We have 2 types of datatypes for analysis.
    - Numerical: When we have only one numerical column, we plot the `Histogram or KDE plot`.
    - Categorical: When we have only one categorical column, we plot `Count Plot`.

2. `Bivariate Analysis`: When we take two columns for analysis, it is called bivariate analysis.
   - When we have two numerical column, we plot the `scatter plot`.
   - When we have one numerical and one categorical column, we plot `Bar plot`.

3. `Multivariate Analysis`: When we take more than two columns for analysis, it is called multivariate analysis.
   - In this case, we are going to create `Pair plot with HUE Parameter`. (Not recommended)
   - In multivariate analysis, we create `Cross tab or PivotÂ table`.
